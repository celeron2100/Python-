### WordNet是一個由普林斯頓大學認識科學實驗室在心理學教授喬治·A·米勒的指導下建立和維護的英語字典。開發工作從1985年開始，從此以後該項目接受了超過300萬美元的資助（主要來源於對機器翻譯有興趣的政府機構）。
### 由於它包含了語義信息，所以有別於通常意義上的字典。WordNet根據詞條的意義將它們分組，每一個具有相同意義的字條組稱為一個synset（同義詞集合）。WordNet為每一個synset提供了簡短，概要的定義，並記錄不同synset之間的語義關係。


### synset是NLTK下的WordNet模組使用的文字展示格式。格式結構組成由"< lemma >.< pos >.< number >"構成，< lemma > 是文字形態學「詞幹」 ， < pos > 是文字的詞性，也是模組中的可用屬性之一 ， < number > 是文字的相關聯程度，由0開始記數。


### WordNet的開發有兩個目的：
### 它既是一個字典，又是一個辭典，它比單純的辭典或詞典都更加易於使用。
### 支持自動的文本分析以及人工智慧應用。
### WordNet的資料庫及相應的軟體工具的發放遵照BSD許可證書，可以自由的下載和使用，亦可在線查詢和使用。

### 在WordNet中，名詞，動詞，形容詞和副詞各自被組織成一個同義詞的網絡，每個同義詞集合都代表一個基本的語義概念，並且這些集合之間也由各種關係連接。

### WordNet的API中，使用了幾種特殊的代號來表示名詞，動詞，形容詞和副詞這些英文詞性，在WordNet中這樣的詞性標註則稱為< pos > 。 

### 以下列舉幾個常見的 < pos > 代號
#### ADJ = 'a'

#### ADJ_SAT = 's'

#### ADV = 'r'

#### NOUN = 'n'

#### VERB = 'v'

### ADJ 與 ADJ_SAT 的差異
Adjectives are arranged in clusters containing head synsets and satellite synsets. Each cluster is organized around antonymous pairs (and occasionally antonymous triplets). The antonymous pairs (or triplets) are indicated in the head synsets of a cluster. Most head synsets have one or more satellite synsets, each of which represents a concept that is similar in meaning to the concept represented by the head synset. One way to think of the adjective cluster organization is to visualize a wheel, with a head synset as the hub and satellite synsets as the spokes. Two or more wheels are logically connected via antonymy, which can be thought of as an axle between the wheels.

from http://wordnet.princeton.edu/wordnet/man/wngloss.7WN.html

#### 若不指定要顯示的pos，則預設情況下會顯示所有pos詞性的Synset。

```python
synse = wn.synsets('good')
print(synse)
```

#### 如果加入了 "pos = "參數，就可以指定要顯示的pos詞性Synset。

```python
synse = wn.synsets('good',pos=wn.NOUN)
print(synse)
```

```python
synse = wn.synsets('good',pos=wn.ADJ)
print(synse)
```

```python
synse = wn.synsets('good',pos=wn.ADJ_SAT)
print(synse)
```

```python
synse = wn.synsets('good',pos=wn.ADV)
print(synse)
```

### pos參數部分可以輸入單一的pos也可以輸入一組的pos tuple來顯示中種pos標註

```python
synse = wn.synsets('good',pos=(wn.ADV,wn.NOUN))
print(synse)
```

### 基礎英文語句的斷詞，將英文語句切割成一單詞組成的串列(List)

```python
try_sentence = "The quick brown fox jumps over the lazy dog"
try_sentence = try_sentence.split()
print(try_sentence)
```

### 使用簡單的迴圈就可以把英文語句中所有英文單字的全部pos標註的synset顯示出來。

```python
sentiment_score_list = []
for senti_word_text in try_sentence:
    synsets_adj = wn.synsets(senti_word_text,pos=wn.ADJ)
    synsets_adv = wn.synsets(senti_word_text,pos=wn.ADV)
    synsets_verb = wn.synsets(senti_word_text,pos=wn.VERB)
    synsets_noun = wn.synsets(senti_word_text,pos=wn.NOUN)
    temp = synsets_adj+synsets_adv+synsets_verb+synsets_noun
    sentiment_score_list = sentiment_score_list + temp
print(sentiment_score_list)    
```

### SentiWordNet是一個用來進行意見挖掘(Opinion mininig)的詞彙庫資源，SentiWordNet對WordNet中的每一組同義詞集分配三個情緒分數(Sentiment Score)，每一組同義詞都會有正向(positivity),負向(negativity),客觀(objectivity)三種情緒極性分數。

### 目前的SentiWordNet的版本是3.0版，也是基於WordNet 3.0版本擴充而成。

```python
print("Synset 一共有 {} 組".format(len(wn.synsets("apple"))))
print(wn.synsets("apple"))  #在查詢WordNet的synset時，回傳的結果是用串列(List)格式
```

```python
print(type(wn.synsets("apple")[0]))  #字串中的元素則是NLTK特有的類別格式，並不利於我們在程式中應用
```

```python
print(type(wn.synsets("apple")[0].name()))
print(wn.synsets("apple")[0].name())  # 加入.name()屬性，可以觀看WordNet中純字串(string)型態的synset
```

```python
wn.synsets("apple")[1].name()
```

#### 因為apple這個英文單詞在WordNet中只有定義兩組synset，所以查詢所以超過詞彙長度的話就會出現Index Error

```python
try:
    wn.synsets("apple")[2].name()
except IndexError:
    print("IndexError: list index out of range")
```

```python
print([swn.senti_synset(word.name()) for word in wn.synsets("apple")])
```

```python
for senti_word_text in sentiment_score_list:
    senti_synset = swn.senti_synset(senti_word_text.name())
    print(senti_synset.synset)
```

```python
for senti_word_text in sentiment_score_list:
    senti_synset = swn.senti_synset(senti_word_text.name())
    print("| {0}的正向情緒分數:{1} | 負向情緒分數:{2} | 客觀情緒分數:{3} |".format(senti_synset.synset.name(),
                                    senti_synset.pos_score(),
                                   senti_synset.neg_score(),
                                   senti_synset.obj_score()))
```

```python
parse_word = swn.senti_synsets('sweet', 'a')

parse_word_list = list(parse_word)[0]

print(parse_word_list)

print("parse_word_list.pos_score():{0}".format(parse_word_list.pos_score()))
print("parse_word_list.neg_score():{0}".format(parse_word_list.neg_score()))
print("parse_word_list.obj_score():{0}".format(parse_word_list.obj_score()))
```

```python
# -*- coding: utf-8 -*-
from nltk import tokenize
from __future__ import print_function
from nltk.sentiment.vader import SentimentIntensityAnalyzer


sentences =["The quick brown fox jumps over the lazy dog."]

sid = SentimentIntensityAnalyzer()

for sentence in sentences:
    print(sentence)
    ss = sid.polarity_scores(sentence)
    for k in sorted(ss):
            print("{0}:{1}".format(k, ss[k]),end=" ")
print()
```
