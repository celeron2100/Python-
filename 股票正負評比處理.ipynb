```python
import jieba
jieba.load_userdict("jieba_custom_dict.txt")
```

```python
import pandas as pd
```

```python
pro_noun = pd.read_csv("股市專有名詞.csv",sep=',',header=None,names=["name","context"])
```

```python
pos_corpus_file = open("Chinese_sentiment_corpus/positive_corpus.txt","r",encoding = 'utf8')
pos_corpus_read = pos_corpus_file.read()
pos_corpus_read=pos_corpus_read.split("\n")

neg_corpus_file = open("Chinese_sentiment_corpus/negative_corpus.txt","r",encoding = 'utf8')
neg_corpus_read = neg_corpus_file.read()
neg_corpus_read=neg_corpus_read.split("\n")

print("正向評比詞彙庫總辭數 : ",len(list(set(pos_corpus_read)))," 個詞彙")
print("負向評比詞彙庫總辭數 : ",len(list(set(neg_corpus_read)))," 個詞彙")


pos_corpus_read=list(set(pos_corpus_read))
neg_corpus_read=list(set(neg_corpus_read))

print("正向詞彙表前20個詞彙預覽\n")
print(pos_corpus_read[:20])
print("\n\n")
print("負向詞彙表前20個詞彙預覽\n")
print(neg_corpus_read[:20])
```

```python
pro_noun["name"][0]
```

```python
counter = 0
for context_only in pro_noun.context:
    pos_sentiment_score = 0
    neg_sentiment_score = 0
    neutral_sentiment_score = 0
    print("|*專有名詞",counter+1,"號 : ",pro_noun["name"][counter])
    counter +=1
    print("|*名詞解釋: ",context_only)
    try:
        cut_sentence = ",".join(jieba.cut_for_search(context_only))
        print("|*精細斷詞: ",cut_sentence)
        for j in cut_sentence:
            if j in set(pos_corpus_read):
                pos_sentiment_score +=1
            if j in set(neg_corpus_read):
                neg_sentiment_score -=1
            else:
                neutral_sentiment_score +=1
        print("|*估計情緒分數==>  正向情緒詞比例: ",
              pos_sentiment_score," 負向情緒詞比例: ",
              neg_sentiment_score#," 中性情緒詞比例: ",
              #neutral_sentiment_score/len(cut_sentence)
             )
        print("|\n|","___"*45)
        print("|")
    except AttributeError:
        pass
    continue
```
