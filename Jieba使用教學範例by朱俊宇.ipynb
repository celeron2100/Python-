```python
# -*- coding: utf-8 -*-

#原教學文件是使用python3
#本範例使用python2.7
from nltk import *
from __future__ import division
from sklearn.feature_extraction.text import CountVectorizer
from __future__ import unicode_literals
import jieba
import jieba.analyse as als
import jieba.posseg as psg
from __future__ import unicode_literals
from sklearn import feature_extraction
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#詞性標記

word_01 = psg.cut("人工智慧 指 的 是 由 人工 製 造 出 來 的 系 統 所 表 現 出 來 的 智慧 ， 通常 是 指透過 普通 電腦 實現 的 智慧 。")

for i,j in word_01:
    print "%s  %s  "%(i,j)
```

```python
#tokenize練習

tokenize_01 = jieba.tokenize(u"標記的練習範例") #記得要加小u，以unicode輸入

for i in tokenize_01:
    print "word %s     start:%d     end:%d"%(i[0],i[1],i[2])  #此句只能使用英文格式!
```

```python
#搜索模式

search_01 = jieba.tokenize(u"搜索模式練習",mode="search")

for i in search_01:
    print "word %s     start:%d     end:%d"%(i[0],i[1],i[2])
```

```python
#cut練習
sentence_01 = jieba.cut("人工智慧來勢洶洶 對人類是福是禍？")

print "/".join(sentence_01)

sentence_01 = jieba.cut("人工智慧來勢洶洶 對人類是福是禍？",cut_all= True)

print "/".join(sentence_01)

sentence_01 = jieba.cut_for_search("人工智慧來勢洶洶 對人類是福是禍？")

print "/".join(sentence_01)
```

```python
article_list = []

file_01 = open("python_article.txt","r")
for i in file_01:
    article_list.append(" ".join(jieba.cut(i.decode("utf-8").split(",")[0],cut_all=False)))

vector_01 = TfidfVectorizer()
tfidf_01 = vector_01.fit_transform(article_list)
print tfidf_01.shape
```

```python
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(article_list)
print X_train_counts.shape

```

```python
for i in article_list:
    print i.encode('utf-8')
```

```python
#計算TF-IDF
#記得安裝ifidf套件pip install tfidf


words_01 = vector_01.get_feature_names()

for i in words_01:
    print i
```

```python
for i in count_vect.vocabulary_:
    print i
```

```python
fq = nltk.FreqDist(count_vect.vocabulary_)
total =0
counter =0
for i in words_01:
    print i,fq.freq(i)
    total += fq.freq(i)
    counter += 1
    
    
print "\n\n總和",total,"\n\n詞彙數",counter
```
