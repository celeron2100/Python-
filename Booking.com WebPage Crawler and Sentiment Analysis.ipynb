```python
to_calculate_clean_effect =[]
original_length=0
cleaned_punctuations_length=0
cleaned_stopword_length=0
punctuation=['(', ')', '?', ':', ';', ',', '.', '!', '/', '"', "'",'',
             '','\\','[','|','','@',']','','','','','','','',' ','%',
             '-','1','2','3','4','5','6','7','8','9','0',' ','\t','','a', 'b', 'c',
             'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r',
             's', 't', 'u', 'v', 'w', 'x', 'y', 'z','$','.','-']

def preprocess(sentence):
    sentence = sentence.lower()
    global original_length
    original_length += len(word_tokenize(sentence))
    print("\n:",len(word_tokenize(sentence))," \nOrginal Sentence : \n"," ".join(word_tokenize(sentence)))
    
    tokenizer = RegexpTokenizer(r'\w+')
    tokens = tokenizer.tokenize(sentence)
    tokens = [w for w in tokens if not w in punctuation]
    global cleaned_punctuations_length
    cleaned_punctuations_length += len(tokens)
    print("\n:",len(tokens),"\nClean the Punctuations : \n"," ".join(tokens))
    
    filtered_words = [w for w in tokens if not w in stopwords.words('english')]
    print("\n:",len(filtered_words),"\nClean the Stopwords : \n"," ".join(filtered_words))
    global cleaned_stopword_length
    cleaned_stopword_length += len(filtered_words)
    print()
    return filtered_words

for i in list(pandas_dataframe_02_Content):
    preprocess(i)
    tmp = word_tokenize(i)
    original_length +=len(tmp)
    to_calculate_clean_effect.append(tmp)
    
```

```python
def easy_sentiment_score(senti_word_0001):
    total_sc = 0
    global one_sentence_score
    synsets_adj = wn.synsets(senti_word_0001,pos=wn.ADJ)
    synsets_adv = wn.synsets(senti_word_0001,pos=wn.ADV)
    synsets_verb = wn.synsets(senti_word_0001,pos=wn.VERB)
    synsets_noun = wn.synsets(senti_word_0001,pos=wn.NOUN)
    for arr_length in range(len(synsets_adj)):
        target_name = synsets_adj[arr_length]
        single_sentiment_score=swn.senti_synset(target_name.name())
        pos_s=single_sentiment_score.pos_score()
        neg_s=single_sentiment_score.neg_score()
        total_sc = pos_s-neg_s
        one_sentence_score += total_sc
        #print("The Adj  Polarity Score :  %2.4f   -  %2.4f    =  %2.4f "%(pos_s,neg_s,total_sc))
    for arr_length in range(len(synsets_noun)):
        target_name = synsets_noun[arr_length]
        single_sentiment_score=swn.senti_synset(target_name.name())
        pos_s=single_sentiment_score.pos_score()
        neg_s=single_sentiment_score.neg_score()
        total_sc = pos_s-neg_s
        one_sentence_score += total_sc
        #print("The Noun Polarity Score :  %2.4f   -  %2.4f    =  %2.4f "%(pos_s,neg_s,total_sc))
    for arr_length in range(len(synsets_adv)):
        target_name = synsets_adv[arr_length]
        single_sentiment_score = swn.senti_synset(target_name.name())
        pos_s=single_sentiment_score.pos_score()
        neg_s=single_sentiment_score.neg_score()
        total_sc = pos_s-neg_s
        one_sentence_score += total_sc
        #print("The Adv  Polarity Score :  %2.4f   -  %2.4f    =  %2.4f "%(pos_s,neg_s,total_sc))
    for arr_length in range(len(synsets_verb)):
        target_name = synsets_verb[arr_length]
        single_sentiment_score = swn.senti_synset(target_name.name())
        pos_s=single_sentiment_score.pos_score()
        neg_s=single_sentiment_score.neg_score()
        total_sc = pos_s-neg_s
        one_sentence_score += total_sc
        #print("The Verb Polarity Score :  %2.4f   -  %2.4f    =  %2.4f "%(pos_s,neg_s,total_sc))    
```

```python
sentence_index_count = 1
sentence_context_list = []
sentenct_polarity = []
sentenct_score = []

for positive_review in list(pandas_dataframe_02_Content):
    one_sentence_score = 0
    print("\ncurrent sentence number : ",sentence_index_count)
    sentence_list = preprocess(positive_review)
    #print("\nSentence : ",sentence_list)
    sentence_index_count +=1
    for j in sentence_list:
        
        try:
            easy_sentiment_score(j)
        except IndexError:
            pass
        continue
    print("One sentence score is ",one_sentence_score,"\n\n")
    sentenct_score.append(one_sentence_score)
    sentence_context_list.append(" ".join(sentence_list))
    if one_sentence_score == 0:
        sentenct_polarity.append("Neutral")
    if one_sentence_score > 0:
        sentenct_polarity.append("Positive")
    if one_sentence_score < 0:
        sentenct_polarity.append("Negative")
    
```

```python
# coding=UTF-8
import nltk
from nltk.corpus import brown

brown_train = brown.tagged_sents(categories='news')
regexp_tagger = nltk.RegexpTagger(
    [(r'^-?[0-9]+(.[0-9]+)?$', 'CD'),
     (r'(-|:|;)$', ':'),
     (r'\'*$', 'MD'),
     (r'(The|the|A|a|An|an)$', 'AT'),
     (r'.*able$', 'JJ'),
     (r'^[A-Z].*$', 'NNP'),
     (r'.*ness$', 'NN'),
     (r'.*ly$', 'RB'),
     (r'.*s$', 'NNS'),
     (r'.*ing$', 'VBG'),
     (r'.*ed$', 'VBD'),
     (r'.*', 'NN')
     ])

unigram_tagger = nltk.UnigramTagger(brown_train, backoff=regexp_tagger)
bigram_tagger = nltk.BigramTagger(brown_train, backoff=unigram_tagger)
cfg = {}
cfg["NNP+NNP"] = "NNP"
cfg["NN+NN"] = "NNI"
cfg["NNI+NN"] = "NNI"
cfg["JJ+JJ"] = "JJ"
cfg["JJ+NN"] = "NNI"

class NPExtractor(object):
    def __init__(self, sentence):
        self.sentence = sentence
    def tokenize_sentence(self, sentence):
        tokens = nltk.word_tokenize(sentence)
        return tokens
    def normalize_tags(self, tagged):
        n_tagged = []
        for t in tagged:
            if t[1] == "NP-TL" or t[1] == "NP":
                n_tagged.append((t[0], "NNP"))
                continue
            if t[1].endswith("-TL"):
                n_tagged.append((t[0], t[1][:-3]))
                continue
            if t[1].endswith("S"):
                n_tagged.append((t[0], t[1][:-1]))
                continue
            n_tagged.append((t[0], t[1]))
        return n_tagged
    def extract(self): 
        tokens = self.tokenize_sentence(self.sentence)
        tags = self.normalize_tags(bigram_tagger.tag(tokens))
        merge = True
        while merge:
            merge = False
            for x in range(0, len(tags) - 1):
                t1 = tags[x]
                t2 = tags[x + 1]
                key = "%s+%s" % (t1[1], t2[1])
                value = cfg.get(key, '')
                if value:
                    merge = True
                    tags.pop(x)
                    tags.pop(x)
                    match = "%s %s" % (t1[0], t2[0])
                    pos = value
                    tags.insert(x, (match, pos))
                    break
        matches = []
        for t in tags:
            if t[1] == "NNP" or t[1] == "NNI":
                matches.append(t[0])
        return matches
```

```python
import string

def The_sentence_is_about(sentence):
    exclude = set(string.punctuation)
    sentence = sentence.lower()
    sentence = ''.join(ch for ch in sentence if ch not in exclude)
    np_extractor = NPExtractor(sentence)
    result = np_extractor.extract()
    #print(type(result))
    print("This sentence is about :  \n%s" % " , ".join(result))
    
    
def The_sentence_is_about_analysis_material(sentence):
    exclude = set(string.punctuation)
    sentence = sentence.lower()
    sentence = ''.join(ch for ch in sentence if ch not in exclude)
    np_extractor = NPExtractor(sentence)
    result = np_extractor.extract()
    return result
```
